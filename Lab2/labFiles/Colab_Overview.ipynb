**Functions**


```
# this function that takes 2 numbers as input and outputs their average value
def avg(x,y):
	print("first input is ",  x )
	print("second input is ", y)
	a = (x + y)/2.0
	print("average is ", a)
	return a

#call the function
avg(200,300)
```
